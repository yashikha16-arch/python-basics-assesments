## Question 9 (Easy)

**Title:** Count Even and Odd Numbers

### Problem Statement:
Take N as input and count how many even and odd numbers are there between 1 and N.

### Approach:
Use two counters and check with modulo operator.


```python
# write your code here
n=int(input())
even_count=0
odd_count=0
for i in range(1,n+1):
  if i%2==0:
    even_count +=1
  else:
    odd_count +=1

print(f"even_count: {even_count}")
print(f"odd_count: {odd_count}")
```

## Question 16 (Hard)

**Title:** Check Prime Number

### Problem Statement:
Take a number as input and check whether it is prime or not.

### Approach:
Loop from 2 to N-1 and use `break` if divisible.


```python
# write your code here
n=int(input("enter the number"))
for i in range(2,n-1):
  if n%i==0:
    print("not prime")
    break
else:
  print("prime number")
```

## Question 17 (Hard)

**Title:** Prime Numbers Between 1 and N

### Problem Statement:
Take N as input and print all prime numbers between 1 and N.

### Approach:
Use nested loops and a flag variable for prime checking.


```python
# write your code here
n=int(input("enter the number"))
for i in range(1,n+1):
  for j in range(2,i):
    if i%j==0:
      break
  else:
    print(i)
```

## Question 18 (Hard)

**Title:** First Number Divisible by 7 and 11

### Problem Statement:
Take N as input and find the first number between 1 and N that is divisible by both 7 and 11. Stop after finding it.

### Approach:
Use logical AND and stop the loop using `break`.


```python
n=int(input("enter the number"))
for i in range(1,n+1):
  if i % 7 == 0 and i % 11 == 0:
    print(i)
    break
```
